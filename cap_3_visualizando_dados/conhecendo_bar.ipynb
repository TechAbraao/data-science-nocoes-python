```python
# Importando o 'pyplot' do 'matplotlib'
from matplotlib import pyplot as plt
```

```python
nomes = ["Abra√£o", "Anthony", "Lucas", "Henry", "Amorim"]
idades = [22, 19, 19, 18, 19]
plt.bar(range(len(nomes)), idades) # o 'plt.bar' √© utilizado para criar gr√°ficos de barra
# isto √©, gr√°ficos de barras s√£o utilizados para compara√ß√£o 
plt.title("Nomes")
plt.ylabel("Idades")
plt.xticks(range(len(nomes)), nomes)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()], histogram.values(), width=8)
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decil")
plt.ylabel("# de Alunos")
plt.title("Distribui√ß√£o das Notas do Teste 1")
plt.show()
```

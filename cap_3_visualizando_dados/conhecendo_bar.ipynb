```python
# Importando o 'pyplot' do 'matplotlib'
from matplotlib import pyplot as plt
```

```python
nomes = ["Abraão", "Anthony", "Lucas", "Henry", "Amorim"]
idades = [22, 19, 19, 18, 19]
plt.bar(range(len(nomes)), idades) # o 'plt.bar' é utilizado para criar gráficos de barra
# isto é, gráficos de barras são utilizados para comparação 
plt.title("Nomes")
plt.ylabel("Idades")
plt.xticks(range(len(nomes)), nomes)
```

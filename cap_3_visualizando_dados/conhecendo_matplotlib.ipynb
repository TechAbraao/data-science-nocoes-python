```python
# Importando o 'pyplot' do 'matplotlib'
from matplotlib import pyplot as plt
```

```python
# Dados que iremos analisar
anos = [1950, 1960, 1980, 1990, 2000, 2010]
bilhoes_dolares = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7]
```

```python
plt.plot(anos, bilhoes_dolares, color='green', marker='o', linestyle='solid')
plt.title("PIB Nominal")
plt.ylabel("Bilhões de Doláres")
plt.show()
```

```python
plot_blue = plt.plot(anos, bilhoes_dolares, color='blue', marker="o", linestyle='solid')
plot_blue
```

```python
# Base de Dados para o criar um gráfico de linhas no matplotlib
altura = [1.60, 1.65, 1.70, 1.75]
idade = [14, 16, 18, 20]
grafico_altura_por_idade = plt.plot(idade, altura, marker="o") # Coordenadas X e Y (x, y)
plt.title("Anos de Idade")
plt.ylabel("Altura em centrímetros")
grafico_altura_por_idade
```
